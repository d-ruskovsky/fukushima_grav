## Homogénna hustota

Prizma jedna

```python
# všetkému som dal príponu L% , pre lineárnu hustotu
Hpoint =  [1,1,20]           # m
Hprism =  [0,10,0,10,0,20]     # m
Hrho_s =  3                    # kg/m3/m

HV, Hg, HG = Fukushima(Hprism,Hpoint,Hrho_s,0,0)

print("Potenciál V = ",HV,"\n")
print("Zrýchlenie g = ",Hg,"\n")
print("Tenzor G = ",HG,"\n")
```

Prizma rozdelená

```python
dZ = 1      # vertikálny krok rozdelenia prizmy
z_min, z_max = Hprism[4], Hprism[5] # začiatok a koniec cyklu

# inicializácia nulových hodnot pre sčítavané veličiny 
HV2 = 0
Hg2 = 0
HG2 = 0

# cyklus pre výpočet jednej rozdelenej prizmy
for z1 in range(z_min, z_max, dZ):
    z2 = z1 + dZ
    subprism = [Hprism[0],Hprism[1],Hprism[2], Hprism[3], z1, z2]

    V2, g2, G2 = Fukushima(subprism, Hpoint, Hrho_s,0,0)

    HV2 += V2
    Hg2 += g2
    HG2 += G2

print("Potenciál V = ",HV2,"\n")
print("Zrýchlenie g = ",Hg2,"\n")
print("Tenzor G = ",HG2,"\n")
```

Rozdiel

```python
dHV = HV - HV2
dHg = Hg - Hg2
dHG = HG - HG2

print("Rozdiel potenciálu V = ",dHV,"\n")
print("Rozdiel zrýchlenia g = ",dHg,"\n")
print("Rozdiel tenzoru G = ",dHG,"\n")
```

## Lineárna hustota

prizma jedna

```python
# všetkému som dal príponu L% , pre lineárnu hustotu
Lpoint =  [1,1,20]           # m
Lprism =  [0,10,0,10,0,20]     # m
Lrho_s =  3                    # kg/m3
Lrho_gradient = 1               # kg/m3/m

LV, Lg, LG = Fukushima(Lprism,Lpoint,Lrho_s,Lrho_gradient,1)

print("Potenciál V = ",LV,"\n")
print("Zrýchlenie g = ",Lg,"\n")
print("Tenzor G = ",LG,"\n")
```

prizma rozdelená

```python
dZ = 1      # vertikálny krok rozdelenia prizmy
z_min, z_max = Lprism[4], Lprism[5] # začiatok a koniec cyklu

# inicializácia nulových hodnot pre sčítavané veličiny 
LV2 = 0
Lg2 = 0
LG2 = 0

# cyklus pre výpočet jednej rozdelenej prizmy
for z1 in range(z_min, z_max, dZ):
    z2 = z1 + dZ
    subprism = [Lprism[0],Lprism[1],Lprism[2], Lprism[3], z1, z2]

    V2, g2, G2 = Fukushima(subprism, Lpoint, Lrho_s + 1,0,0)

    LV2 += V2
    Lg2 += g2
    LG2 += G2

print("Potenciál V = ",HV2,"\n")
print("Zrýchlenie g = ",Hg2,"\n")
print("Tenzor G = ",HG2,"\n")
```

rozdiel

```python
dLV = LV - LV2
dLg = Lg - Lg2
dLG = LG - LG2

print("Rozdiel potenciálu V = ",dLV,"\n")
print("Rozdiel zrýchlenia g = ",dLg,"\n")
print("Rozdiel tenzoru G = ",dLG,"\n")
```
