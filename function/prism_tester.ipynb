Vytvorenie prizmy 10x10x10m s danou hustotou 3 kg/m^3/m

```python
from fukushima import *
```

## Lineárna hustota

Prizma jedna

```python
# všetkému som dal príponu L% , pre lineárnu hustotu
Lpoint =  [5000,5000,6000]           # m
Lprism =  [0,1000,0,1000,0,5000]     # m
Lrho_s =  15000                    # kg/m3/m

LV, Lg, LG = Fukushima(Lprism,Lpoint,Lrho_s,0,0)

print("Potenciál V = ",LV,"\n")
print("Zrýchlenie g = ",Lg,"\n")
print("Tenzor G = ",LG,"\n")
```

Prizma rozdelená

```python
dZ = 10      # vertikálny krok rozdelenia prizmy
z_min, z_max = Lprism[4], Lprism[5] # začiatok a koniec cyklu

# inicializácia nulových hodnot pre sčítavané veličiny 
LV2 = 0
Lg2 = 0
LG2 = 0

# cyklus pre výpočet jednej rozdelenej prizmy
for z1 in range(z_min, z_max, dZ):
    z2 = z1 + dZ
    subprism = [Lprism[0],Lprism[1],Lprism[2], Lprism[3], z1, z2]

    V2, g2, G2 = Fukushima(subprism, Lpoint, Lrho_s,0,0)

    LV2 += V2
    Lg2 += g2
    LG2 += G2

print("Potenciál V = ",LV2,"\n")
print("Zrýchlenie g = ",Lg2,"\n")
print("Tenzor G = ",LG2,"\n")
```

Rozdiel

```python
dLV = LV - LV2
dLg = Lg - Lg2
dLG = LG - LG2

print("Rozdiel potenciálu V = ",dLV,"\n")
print("Rozdiel zrýchlenia g = ",dLg,"\n")
print("Rozdiel tenzoru G = ",dLG,"\n")
```
